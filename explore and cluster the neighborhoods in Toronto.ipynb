### Import libary

```python
import requests
import numpy as np
import pandas as pd
```

### Get url text infomation use requests function

```python
#define the target URL 
TargetURL="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
#use requests to get text file from website
website_url = requests.get(TargetURL).text
website_url
```

Extract table from html text file use BeautifulSoup

```python
# use BeautifulSoup to get process on web text
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
print(My_table)
```

```python
# use findall to ge the lists includine 'td'
links=My_table.findAll('td')
links
```

### Import table info into datafrme and clean the table

reshape the list 

```python
table=np.array(links).reshape((len(links)//3,3))
Table=pd.DataFrame(table,columns=["PostalCode","Borough",'Neighborhood'])
Table.astype(str)
Table.head()
```

clean the table , revome unuseful tag

```python
#remove '<td>' and '</td>' and '</a>' and']'
for col in Table.columns:
    Table[col]=Table[col].astype(str).str.replace('<td>','').str.replace('</td>','').str.replace('</a>','').str.replace('\n','').str.replace(']','')
Table.head()
```

```python
#remove on 'Borough ' and 'Neighborhood'
Table['Borough']=Table['Borough'].astype(str).str.split('title=').str[-1].str.split('>').str[-1]
Table['Neighborhood']=Table['Neighborhood'].astype(str).str.split('title=').str[-1].str.split('>').str[-1]
Table.head()
```

```python
#replace 
Table.replace('Not assigned',np.nan,inplace=True)
Table.head()
```

```python
#auto fill like cell (1,7) 
#print(Table[(~Table.Borough.isnull()) & (Table.Neighborhood.isnull())]) #there is only one so fill it manually

Table.Neighborhood[8]=Table.Borough[8]

Table.iloc[8,:]
```

```python
#dropna
Table.dropna(inplace=True)
Table.reset_index(drop=True,inplace=True)
Table.head()
```
